```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('example_dataset.csv')
df = pd.DataFrame(data)
```

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(7,5))
gender_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Distribution of Gender in the Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['Age'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Age in the Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
